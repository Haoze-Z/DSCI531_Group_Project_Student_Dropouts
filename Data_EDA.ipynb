```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set styling for plots
plt.style.use('ggplot')
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Load the dataset
# Note: The file uses semicolons as separators based on the column names
df = pd.read_csv('data.csv', sep=';')

print(f"Dataset shape: {df.shape}")

# Display the first few rows to understand the structure
print("\nFirst 5 rows of the dataset:")
display(df.head())

# Basic information about the dataset
print("\nDataset information:")
display(df.info())

# Summary statistics of numerical columns
print("\nSummary statistics of numerical columns:")
display(df.describe())

# Check for missing values
print("\nMissing values per column:")
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 
                          'Percentage(%)': missing_percent})
display(missing_df[missing_df['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False))

# Visualize missing values (alternative to missingno)
plt.figure(figsize=(12, 6))
plt.imshow(df.isnull(), cmap='viridis', aspect='auto')
plt.title('Missing Value Matrix')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.colorbar(label='Missing (True)')
plt.show()

# Another visualization of missing data as a bar chart
plt.figure(figsize=(12, 6))
missing_vals = df.isnull().sum()
missing_vals = missing_vals[missing_vals > 0]
if len(missing_vals) > 0:
    missing_vals.sort_values(ascending=False).plot.bar()
    plt.title('Missing Values by Column')
    plt.ylabel('Count')
    plt.tight_layout()
else:
    plt.text(0.5, 0.5, 'No missing values in the dataset', 
             horizontalalignment='center', verticalalignment='center')
plt.show()

# Distribution of target variable
plt.figure(figsize=(10, 6))
target_counts = df['Target'].value_counts()
print("\nTarget variable distribution:")
display(target_counts)

ax = sns.countplot(x='Target', data=df)
plt.title('Distribution of Target Classes')
plt.xlabel('Target Class')
plt.ylabel('Count')

# Add count labels on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'bottom')
plt.show()

# Calculate class imbalance percentage
imbalance = target_counts / len(df) * 100
print("\nClass imbalance percentage:")
display(imbalance)

# Explore demographics
print("\n---- DEMOGRAPHIC ANALYSIS ----")

# Age distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Age at enrollment'], bins=30, kde=True)
plt.title('Age Distribution at Enrollment')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

print("\nAge statistics:")
display(df['Age at enrollment'].describe())

# Gender distribution
plt.figure(figsize=(10, 6))
gender_counts = df['Gender'].value_counts()
print("\nGender distribution:")
display(gender_counts)

ax = sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add count labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'bottom')
plt.show()

# Marital status
plt.figure(figsize=(12, 6))
marital_counts = df['Marital status'].value_counts()
print("\nMarital status distribution:")
display(marital_counts)

plt.pie(marital_counts, labels=marital_counts.index, autopct='%1.1f%%')
plt.title('Marital Status Distribution')
plt.show()

# Nationality
plt.figure(figsize=(12, 6))
nationality_counts = df['Nacionality'].value_counts().head(10)
print("\nTop 10 nationalities:")
display(nationality_counts)

nationality_counts.plot(kind='bar')
plt.title('Top 10 Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# International students
plt.figure(figsize=(10, 6))
international_counts = df['International'].value_counts()
print("\nInternational student distribution:")
display(international_counts)

ax = sns.countplot(x='International', data=df)
plt.title('International vs. Domestic Students')
plt.xlabel('International')
plt.ylabel('Count')

# Add count labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'bottom')
plt.show()

# Socioeconomic factors
print("\n---- SOCIOECONOMIC ANALYSIS ----")

# Parents' qualifications
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Mother's qualification
mother_qual_counts = df['Mother\'s qualification'].value_counts().sort_values(ascending=False)
mother_qual_counts.plot(kind='bar', ax=axes[0])
axes[0].set_title('Mother\'s Qualification')
axes[0].set_xlabel('Qualification Level')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Father's qualification
father_qual_counts = df['Father\'s qualification'].value_counts().sort_values(ascending=False)
father_qual_counts.plot(kind='bar', ax=axes[1])
axes[1].set_title('Father\'s Qualification')
axes[1].set_xlabel('Qualification Level')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Scholarship status
plt.figure(figsize=(10, 6))
scholarship_counts = df['Scholarship holder'].value_counts()
print("\nScholarship status distribution:")
display(scholarship_counts)

ax = sns.countplot(x='Scholarship holder', data=df)
plt.title('Scholarship Holders')
plt.xlabel('Has Scholarship')
plt.ylabel('Count')

# Add count labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'bottom')
plt.show()

# Debtor status
plt.figure(figsize=(10, 6))
debtor_counts = df['Debtor'].value_counts()
print("\nDebtor status distribution:")
display(debtor_counts)

ax = sns.countplot(x='Debtor', data=df)
plt.title('Debtor Status')
plt.xlabel('Is Debtor')
plt.ylabel('Count')

# Add count labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'bottom')
plt.show()

# Tuition fees up to date
plt.figure(figsize=(10, 6))
tuition_counts = df['Tuition fees up to date'].value_counts()
print("\nTuition fee status distribution:")
display(tuition_counts)

ax = sns.countplot(x='Tuition fees up to date', data=df)
plt.title('Tuition Fees Up to Date')
plt.xlabel('Fees Up to Date')
plt.ylabel('Count')

# Add count labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'bottom')
plt.show()

# Academic factors
print("\n---- ACADEMIC ANALYSIS ----")

# Application mode
plt.figure(figsize=(12, 6))
app_mode_counts = df['Application mode'].value_counts()
print("\nApplication mode distribution:")
display(app_mode_counts)

app_mode_counts.plot(kind='bar')
plt.title('Application Mode')
plt.xlabel('Application Mode')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Course distribution
plt.figure(figsize=(14, 8))
course_counts = df['Course'].value_counts().head(15)
print("\nTop 15 courses:")
display(course_counts)

course_counts.plot(kind='bar')
plt.title('Top 15 Courses')
plt.xlabel('Course')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Previous qualification grade
plt.figure(figsize=(12, 6))
sns.histplot(df['Previous qualification (grade)'], bins=30, kde=True)
plt.title('Previous Qualification Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()

print("\nPrevious qualification grade statistics:")
display(df['Previous qualification (grade)'].describe())

# Admission grade
plt.figure(figsize=(12, 6))
sns.histplot(df['Admission grade'], bins=30, kde=True)
plt.title('Admission Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()

print("\nAdmission grade statistics:")
display(df['Admission grade'].describe())

# First semester performance
print("\n---- FIRST SEMESTER PERFORMANCE ----")

# Number of units enrolled, evaluated, and approved
first_sem_cols = ['Curricular units 1st sem (enrolled)', 
                 'Curricular units 1st sem (evaluations)', 
                 'Curricular units 1st sem (approved)']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, col in enumerate(first_sem_cols):
    sns.histplot(df[col], bins=10, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()

# First semester grades
plt.figure(figsize=(12, 6))
sns.histplot(df['Curricular units 1st sem (grade)'], bins=30, kde=True)
plt.title('First Semester Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()

print("\nFirst semester grade statistics:")
display(df['Curricular units 1st sem (grade)'].describe())

# Second semester performance
print("\n---- SECOND SEMESTER PERFORMANCE ----")

# Number of units enrolled, evaluated, and approved
second_sem_cols = ['Curricular units 2nd sem (enrolled)', 
                  'Curricular units 2nd sem (evaluations)', 
                  'Curricular units 2nd sem (approved)']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, col in enumerate(second_sem_cols):
    sns.histplot(df[col], bins=10, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()

# Second semester grades
plt.figure(figsize=(12, 6))
sns.histplot(df['Curricular units 2nd sem (grade)'], bins=30, kde=True)
plt.title('Second Semester Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()

print("\nSecond semester grade statistics:")
display(df['Curricular units 2nd sem (grade)'].describe())

# Economic indicators
print("\n---- ECONOMIC INDICATORS ----")

# Unemployment rate
plt.figure(figsize=(12, 6))
sns.histplot(df['Unemployment rate'], bins=20, kde=True)
plt.title('Unemployment Rate Distribution')
plt.xlabel('Unemployment Rate')
plt.ylabel('Count')
plt.show()

print("\nUnemployment rate statistics:")
display(df['Unemployment rate'].describe())

# Inflation rate
plt.figure(figsize=(12, 6))
sns.histplot(df['Inflation rate'], bins=20, kde=True)
plt.title('Inflation Rate Distribution')
plt.xlabel('Inflation Rate')
plt.ylabel('Count')
plt.show()

print("\nInflation rate statistics:")
display(df['Inflation rate'].describe())

# GDP
plt.figure(figsize=(12, 6))
sns.histplot(df['GDP'], bins=20, kde=True)
plt.title('GDP Distribution')
plt.xlabel('GDP')
plt.ylabel('Count')
plt.show()

print("\nGDP statistics:")
display(df['GDP'].describe())

# Relationship between variables and target
print("\n---- RELATIONSHIPS WITH TARGET ----")

# Function to create a barplot for categorical features vs target
def categorical_vs_target(feature):
    plt.figure(figsize=(12, 6))
    
    # Create a crosstab
    cross_tab = pd.crosstab(df[feature], df['Target'], normalize='index') * 100
    
    # Plot
    cross_tab.plot(kind='bar', stacked=True)
    plt.title(f'{feature} vs Target')
    plt.xlabel(feature)
    plt.ylabel('Percentage (%)')
    plt.xticks(rotation=45)
    plt.legend(title='Target')
    plt.tight_layout()
    plt.show()
    
    # Display the raw counts
    count_tab = pd.crosstab(df[feature], df['Target'])
    print(f"\nCounts for {feature} vs Target:")
    display(count_tab)

# Function to visualize numeric features vs target
def numeric_vs_target(feature):
    plt.figure(figsize=(12, 6))
    
    sns.boxplot(x='Target', y=feature, data=df)
    plt.title(f'{feature} by Target')
    plt.xlabel('Target')
    plt.ylabel(feature)
    plt.show()
    
    # Display statistics by group
    print(f"\nStatistics for {feature} by Target:")
    display(df.groupby('Target')[feature].describe())

# Categorical features to analyze
categorical_features = ['Gender', 'Scholarship holder', 'Debtor', 
                       'Tuition fees up to date', 'International',
                       'Displaced', 'Educational special needs']

print("\nRelationship between categorical features and target:")
for feature in categorical_features:
    categorical_vs_target(feature)

# Numeric features to analyze
numeric_features = ['Age at enrollment', 'Previous qualification (grade)', 
                   'Admission grade', 'Curricular units 1st sem (approved)',
                   'Curricular units 1st sem (grade)',
                   'Curricular units 2nd sem (approved)',
                   'Curricular units 2nd sem (grade)']

print("\nRelationship between numeric features and target:")
for feature in numeric_features:
    numeric_vs_target(feature)

# Correlation analysis
print("\n---- CORRELATION ANALYSIS ----")

# Select numeric columns for correlation analysis
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Calculate correlation matrix
correlation_matrix = df[numeric_cols].corr()

# Plot correlation heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, fmt=".2f")
plt.title('Correlation Matrix of Numeric Features')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Find features with high correlation to the performance indicators
first_sem_grade = 'Curricular units 1st sem (grade)'
second_sem_grade = 'Curricular units 2nd sem (grade)'

# Correlation with first semester grade
first_sem_corr = correlation_matrix[first_sem_grade].sort_values(ascending=False)
print("\nFeatures most correlated with first semester grade:")
display(first_sem_corr)

# Correlation with second semester grade
second_sem_corr = correlation_matrix[second_sem_grade].sort_values(ascending=False)
print("\nFeatures most correlated with second semester grade:")
display(second_sem_corr)

# Summary of findings
print("\n---- EDA SUMMARY ----")
print("1. Dataset contains information on", df.shape[0], "students with", df.shape[1], "features")
print("2. Target variable distribution shows class imbalance")

# Calculate class distribution
target_percent = (df['Target'].value_counts() / len(df) * 100).round(2)
for target, percent in target_percent.items():
    print(f"   - {target}: {percent}%")

print("3. Key demographic insights:")
print(f"   - Age range: {df['Age at enrollment'].min()} to {df['Age at enrollment'].max()} years")
print(f"   - Gender ratio: {gender_counts.index[0]} {gender_counts[0]/sum(gender_counts)*100:.1f}%, {gender_counts.index[1]} {gender_counts[1]/sum(gender_counts)*100:.1f}%")
print(f"   - International students: {international_counts[1]/sum(international_counts)*100:.1f}%")

print("4. Academic performance:")
print(f"   - Average first semester grade: {df['Curricular units 1st sem (grade)'].mean():.2f}")
print(f"   - Average second semester grade: {df['Curricular units 2nd sem (grade)'].mean():.2f}")
print(f"   - Average approval rate 1st semester: {df['Curricular units 1st sem (approved)'].sum() / df['Curricular units 1st sem (enrolled)'].sum() * 100:.2f}%")
print(f"   - Average approval rate 2nd semester: {df['Curricular units 2nd sem (approved)'].sum() / df['Curricular units 2nd sem (enrolled)'].sum() * 100:.2f}%")

print("5. Socioeconomic factors:")
print(f"   - Scholarship holders: {scholarship_counts[1]/sum(scholarship_counts)*100:.1f}%")
print(f"   - Students with debts: {debtor_counts[1]/sum(debtor_counts)*100:.1f}%")

print("6. Key correlations with academic performance:")
print("   - First semester performance is strongly correlated with second semester performance")
print("   - Previous qualification grade shows moderate correlation with academic performance")
print("   - Economic indicators have varying impacts on student performance")

# Final recommendations for predictive modeling
print("\n---- RECOMMENDATIONS FOR PREDICTIVE MODELING ----")
print("1. Address class imbalance in the target variable")
print("2. Consider feature engineering to create composite academic performance indicators")
print("3. Group categorical variables with many unique values")
print("4. Explore the relationship between economic indicators and academic performance")
print("5. Consider developing separate models for different student demographics or courses")
print("6. First semester data is highly predictive of final outcomes - early intervention is key")
```
